```python
import requests  
from bs4 import BeautifulSoup
import pandas as pd  
```

```python
url = "https://learning.suwon.go.kr/lmth/01_lecture01.asp"
page = requests.get(url)  
soup = BeautifulSoup(page.text, 'lxml')
total_pages= soup.select_one('.total em').get_text()
total_pages = int(total_pages)
total = total_pages/10+1
## range자리에 range(1,int(total)+1)  
```

```python
for page_num in range(1,6):
    url ='https://learning.suwon.go.kr/lmth/01_lecture01.asp?block=1&page={0}&s_cate='.format(page_num)
    page = requests.get(url)  
    soup = BeautifulSoup(page.text, 'lxml')  
    
    li_tages = soup.select('.webzinBoardList a strong')  
    title = []  
    for li in li_tages:
        title.append(li.text.strip())  
        
    li_tages = soup.select('li .postInfo dl:nth-child(2) dd')
    target = []  
    for li in li_tages:
        target.append(li.text.strip())  
        
    li_tages = soup.select('u')  
    number = []  
    for li in li_tages:
        number.append(li.text.strip())  
        
    li_tages = soup.select('li .postInfo dl:nth-child(3) dd')  
    date=[] 
    for li in li_tages:
        date.append(li.text.strip())  
```

```python
list1 = list(zip(title, target,number, date))  
df = pd.DataFrame(list1, columns=['title','target','number', 'date'])  
df
```

```python
for i in range(len(df.title.str[0] == '[')):
    df['kind'] = df.title.str.split(']').str[0]
    df['title_2'] = df.title.str.split(']').str[1]  
    df['kind'] = df['kind'].str.strip("[")
df
```

```python
#  number 열 변환하여 수강신청율 열 생성하기  
import re 
df['applicator'] = df.number.str.extract(r'(\d+)')
df['total'] = df['number'].str[-3:-1]
df
```

```python
df['applicator'] = pd.to_numeric(df['applicator'])  
df['total'] = pd.to_numeric(df['total'])   

df['rate'] = df['applicator']/df['total'] 
df['rate'] = df['rate'].round(2)  
df
```

```python
df.sort_values(by=["target",'rate'], ascending = [True,False])
df2 = pd.DataFrame(df.title)
```
