## 데이터 읽어오기

## 데이터 전처리   
2020년도의 데이터를 샘플데이터로 먼저 전처리

### (1) 학습자 -> 학습자 id 이름, 강좌종요일 -> 강좌종료일 변경  

```python
df.rename(columns = {'학습자':'학습자id', '강좌종요일':'강좌종료일'}, inplace = True)  
df
```

### (2) 직업, 강사 열 제거하기 

```python
df.drop(['직업', '강사'], axis = 'columns', inplace = True)  
df
```

### (3) 신청일자, 강좌시작일, 강좌종료일 날짜형으로 변환 

```python
df['신청일자'] = pd.to_datetime(df['신청일자'])  
df['강좌시작일'] = pd.to_datetime(df['강좌시작일'])  
df['강좌종료일'] = pd.to_datetime(df['강좌종료일'])  
df.dtypes
```

### (3-1) 새로운 열 생성 - 신청일자 및 강좌시작일 사이 몇일 간격/ 강좌시작일과 강좌종료일 간격 열도 생성?

```python
df['신청기간'] = df['강좌시작일'] - df['신청일자']    
df['강좌기간'] = df['강좌종료일'] - df['강좌시작일']    
df
```
